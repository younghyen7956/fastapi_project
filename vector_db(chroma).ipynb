```python
import chromadb

client = chromadb.PersistentClient(path="./chroma_db")
```

```python
collection = client.get_or_create_collection(name="construction_new") # Create
# collection = client.get_or_create_collection(name="fix_const") # Create
```

```python
collection.peek() # describe 
```

```python
collection.count() # item 개수
```

```python
client.list_collections() # collection_list 출력
```

```python
client.delete_collection(name="construction_new") # collection 삭제
```

```python
collection.add(
    ids=["id1"],
    embeddings=[[1.1, 2.3, 3.2]],  # 이 차원으로 맞춤
    documents=["새로운 문서"],
    metadatas=[{"info": "초기"}]
) # collection에 데이터 삽입
```

```python
import numpy as np

fake_emb = np.random.rand(3).tolist()

collection.update(
    ids=["id1"],
    embeddings=[fake_emb],
    documents=["수정된 문서 내용"],
    metadatas=[{"source": "update"}]
) # 컬렉션에 데이터 업데이트
```

```python
result = collection.get(ids=["id1"], include=['embeddings', 'documents', 'metadatas']) # 컬렉션 데이터 조회
print(result)
```

```python
collection.delete(
    ids=["id1"]
) # 컬렉션에 데이터 삭제
```
